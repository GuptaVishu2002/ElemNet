```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd
```

```python
# import graph data 
def load_data(csvname):
    # load in data
    data = np.asarray(pd.read_csv(csvname))

    # import data and reshape appropriately
    X = data[:,0:-1]
    y = data[:,-1]
    y.shape = (len(y),1)
    
    # pad data with ones for more compact gradient computation
    #o = np.ones((np.shape(X)[0],1))
    #X = np.concatenate((o,X),axis = 1)
    #X = X.T
    
    return X,y
```

```python
x, y = load_data('result_graph.csv')
```

```python
plt.plot(x, linestyle='solid',label='training set')
plt.plot(y, linestyle='solid',label='evaluation set')
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.legend();
```
